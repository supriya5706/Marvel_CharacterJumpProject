20) superheroes with good alignment,with the distribution of speed


```python
plt.figure(figsize=(10,5))
good_alignment = df[df['Alignment'] == 'good']
sns.histplot(data=good_alignment,x= "Speed", color='green')
plt.title('Distribution of Speed')
plt.xlabel('Speed')
plt.show()

```

22) Distribution of Durability Scores by Alignment


```python
durability_by_alignment = df.groupby('Alignment')['Durability'].mean()
durability_by_alignment.plot(kind='bar')
plt.title(" Durability Score by Alignment")
plt.xlabel("Alignment")
plt.ylabel(" Durability Score")
plt.show()
```

23) Proportion of Heroes and Villains by Alignment


```python
align_counts = df['Alignment'].value_counts()
plt.pie(align_counts.values, labels=align_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Heroes and Villains by Alignment')
plt.show()
```

24) Villains strenght vs durability


```python
bad_alignment = df[df['Alignment'] == 'bad'].head(30)
plt.figure(figsize=(10,5))
plot = sns.barplot(data=bad_alignment, x='Name', y='Durability')
plot.set_xticklabels(labels=bad_alignment['Name'], rotation=45)
plt.title('Villains Strength vs Durability')
plt.xlabel('Strength')
plt.ylabel('Durability')


```
